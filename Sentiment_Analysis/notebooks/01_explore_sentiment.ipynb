#### EDA

```python

stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()

def strip_html_tags(text):
    soup = BeautifulSoup(text, "html.parser")
    [s.extract() for s in soup(["iframe", "script"])]
    stripped_text = soup.get_text()
    stripped_text = re.sub(r"[\r|\n|\r\n]+", " ", stripped_text)
    return stripped_text

def remove_accented_chars(text):
    return unicodedata.normalize("NFKD", text).encode("ascii", "ignore").decode("utf-8", "ignore")

def clean_and_normalize_corpus(docs):
    norm_docs = []
    for doc in tqdm(docs):
        doc = strip_html_tags(doc)
        doc = doc.translate(str.maketrans("\n\t\r", "   "))
        doc = doc.lower()
        doc = remove_accented_chars(doc)
        doc = contractions.fix(doc)
        doc = re.sub(r"http\S+|www\S+|https\S+", '', doc)  # Remove URLs
        doc = re.sub(r'\@\w+|\#','', doc)  # Remove mentions and hashtags
        doc = re.sub(r'[^a-zA-Z0-9\s]', '', doc, flags=re.I|re.A)  # Remove special chars
        doc = re.sub(' +', ' ', doc)  # Extra spaces
        doc = doc.strip()

        # Optional: remove stopwords and lemmatize
        tokens = doc.split()
        tokens = [lemmatizer.lemmatize(w) for w in tokens if w not in stop_words]
        doc = ' '.join(tokens)

        norm_docs.append(doc)
    return norm_docs
```

```python
train_dataset['clean_tweet'] = clean_and_normalize_corpus(train_dataset['tweet'])
test_dataset['clean_tweet'] = clean_and_normalize_corpus(test_dataset['tweet'])
```

### Feature Engineering

```python
# TF-IDF setup: include both unigrams and bigrams, ignore very rare words
tv = TfidfVectorizer(
    use_idf=True,
    sublinear_tf=True,
    min_df=5,
    max_df=1.0,
    ngram_range=(1, 2),
    max_features=10000  # Optional: for speed & memory
)

# Fit on training clean tweets
X_train = tv.fit_transform(train_dataset['clean_tweet'])

# Transform test data using the same vectorizer

X_test = tv.transform(test_dataset['clean_tweet'])

# Encode labels to binary (0 = negative, 1 = positive)
y_train = train_dataset['target'].map({'negative': 0, 'positive': 1}).values
y_test = test_dataset['target'].map({'negative': 0, 'positive': 1}).values
```

#### Model Training- Multinomial Naive Bayes

```python
# Initialize and train the model
nb_model = MultinomialNB(alpha = 1.0, fit_prior= True)
nb_model.fit(X_train, y_train)

# Predict on test data
y_pred = nb_model.predict(X_test)
```

```python
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay
```

```python
# Classification report
print("Classification Report:")
print(classification_report(y_test, y_pred, target_names=['Negative', 'Positive']))

# Confusion matrix
cm = confusion_matrix(y_test, y_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=['Negative', 'Positive'])
disp.plot(cmap='Blues')
plt.title('Confusion Matrix of Multinomial Naive Bayes')
plt.tight_layout()
plt.savefig("confution_matrix_MLb", dpi = 300)
plt.show()
```

#### Apply 5-fold CV

```python
from sklearn.model_selection import cross_val_score
from sklearn.naive_bayes import MultinomialNB

# Perform 5-fold cross-validation on Naive Bayes
nb_cv_scores = cross_val_score(MultinomialNB(alpha=1.0), X_train, y_train, cv=5, scoring='accuracy')

print("Naive Bayes CV Accuracy Scores:", nb_cv_scores)
print("Average CV Accuracy:", np.mean(nb_cv_scores))
```

#### Model 2 Logistic regrssion model 

```python
from sklearn.linear_model import LogisticRegression
```

```python
# Create and train logistic regression model
lr = LogisticRegression(
    penalty='l2',
    max_iter=500,
    C=1.0,
    solver='liblinear',
    random_state=42
)
lr.fit(X_train, y_train)
```

```python
# Predict on test data
y_pred_lr = lr.predict(X_test)

# Classification report
from sklearn.metrics import classification_report, confusion_matrix, ConfusionMatrixDisplay, accuracy_score

print("Logistic Regression - Classification Report:")
print(classification_report(y_test, y_pred_lr, target_names=['Negative', 'Positive']))

# Accuracy, Precision, Recall, F1
metrics_lr = classification_report(y_test, y_pred_lr, target_names=['Negative', 'Positive'], output_dict=True)
accuracy_lr = accuracy_score(y_test, y_pred_lr)
precision_lr = metrics_lr['weighted avg']['precision']
recall_lr = metrics_lr['weighted avg']['recall']
f1_lr = metrics_lr['weighted avg']['f1-score']

print(f"Accuracy:  {accuracy_lr:.4f}")
print(f"Precision: {precision_lr:.4f}")
print(f"Recall:    {recall_lr:.4f}")
print(f"F1-Score:  {f1_lr:.4f}")
```

```python
cm_lr = confusion_matrix(y_test, y_pred_lr)
disp_lr = ConfusionMatrixDisplay(confusion_matrix=cm_lr, display_labels=['Negative', 'Positive'])
disp_lr.plot(cmap='Greens')
plt.title('Confusion Matrix - Logistic Regression')
plt.tight_layout()
plt.savefig('confusion_matrix_logreg.png')
plt.show()
```
